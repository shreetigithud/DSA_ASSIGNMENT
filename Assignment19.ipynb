````python
1. Merge k Sorted Lists**

You are given an array ofÂ `k`Â linked-listsÂ `lists`, each linked-list is sorted in ascending order.

*Merge all the linked-lists into one sorted linked-list and return it.*

**Example 1:**

```
Input: lists = [[1,4,5],[1,3,4],[2,6]]
Output: [1,1,2,3,4,4,5,6]
Explanation: The linked-lists are:
[
  1->4->5,
  1->3->4,
  2->6
]
merging them into one sorted list:
1->1->2->3->4->4->5->6

```

**Example 2:**

```
Input: lists = []
Output: []

```

**Example 3:**

```
Input: lists = [[]]
Output: []

```

**Constraints:**

- `k == lists.length`
- `0 <= k <= 10000`
- `0 <= lists[i].length <= 500`
- `-10000 <= lists[i][j] <= 10000`
- `lists[i]`Â is sorted inÂ **ascending order**.
- The sum ofÂ `lists[i].length`Â will not exceedÂ `10000`.
````

````python
import heapq


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def merge_k_lists(lists):
    if not lists:
        return None

    min_heap = []
    dummy = ListNode(0)
    current = dummy

    # Insert the first element from each list into the min-heap
    for i, lst in enumerate(lists):
        if lst:
            heapq.heappush(min_heap, (lst.val, i))

    while min_heap:
        val, list_index = heapq.heappop(min_heap)
        current.next = ListNode(val)
        current = current.next

        list_node = lists[list_index]
        if list_node.next:
            list_node = list_node.next
            lists[list_index] = list_node
            heapq.heappush(min_heap, (list_node.val, list_index))

    return dummy.next


# Construct the list of linked lists
lists = [
    ListNode(1, ListNode(4, ListNode(5))),
    ListNode(1, ListNode(3, ListNode(4))),
    ListNode(2, ListNode(6))
]

# Merge the k sorted lists
merged_list = merge_k_lists(lists)

# Print the values of the merged list
while merged_list:
    print(merged_list.val, end=" ")
    merged_list = merged_list.next


````

````python
<aside>
ðŸ’¡ 2. **Count of Smaller Numbers After Self**

Given an integer arrayÂ `nums`, returnÂ *an integer array*Â `counts`Â *where*Â `counts[i]`Â *is the number of smaller elements to the right of*Â `nums[i]`.

**Example 1:**

```
Input: nums = [5,2,6,1]
Output: [2,1,1,0]
Explanation:
To the right of 5 there are2 smaller elements (2 and 1).
To the right of 2 there is only1 smaller element (1).
To the right of 6 there is1 smaller element (1).
To the right of 1 there is0 smaller element.

```

**Example 2:**

```
Input: nums = [-1]
Output: [0]

```

**Example 3:**

```
Input: nums = [-1,-1]
Output: [0,0]

```

**Constraints:**

- `1 <= nums.length <= 100000`
- `-10000 <= nums[i] <= 10000`
````

````python
class NumNode:
    def __init__(self, value, index):
        self.value = value
        self.index = index
        self.count = 0


def count_smaller_elements(nums):
    if not nums:
        return []

    num_nodes = [NumNode(nums[i], i) for i in range(len(nums))]
    counts = [0] * len(nums)

    def merge_sort(arr):
        if len(arr) <= 1:
            return arr

        mid = len(arr) // 2
        left = merge_sort(arr[:mid])
        right = merge_sort(arr[mid:])

        merged = []
        i, j = 0, 0

        while i < len(left) and j < len(right):
            if left[i].value > right[j].value:
                merged.append(right[j])
                right[j].count += len(left) - i
                j += 1
            else:
                merged.append(left[i])
                i += 1

        merged.extend(left[i:])
        merged.extend(right[j:])

        return merged

    merge_sort(num_nodes)

    for num_node in num_nodes:
        counts[num_node.index] = num_node.count

    return counts

nums = [5, 2, 6, 1]
smaller_counts = count_smaller_elements(nums)
print(smaller_counts)


````

````python
 3. **Sort an Array**

Given an array of integersÂ `nums`, sort the array in ascending order and return it.

You must solve the problemÂ **without using any built-in**Â functions inÂ `O(nlog(n))`Â time complexity and with the smallest space complexity possible.

**Example 1:**

```
Input: nums = [5,2,3,1]
Output: [1,2,3,5]
Explanation: After sorting the array, the positions of some numbers are not changed (for example, 2 and 3), while the positions of other numbers are changed (for example, 1 and 5).

```

**Example 2:**

```
Input: nums = [5,1,1,2,0,0]
Output: [0,0,1,1,2,5]
Explanation: Note that the values of nums are not necessairly unique.

```

**Constraints:**

- `1 <= nums.length <= 5 * 10000`
- `-5 * 104 <= nums[i] <= 5 * 10000`
</aside>
````

````python
def quicksort(nums, start, end):
    if start >= end:
        return

    pivot = nums[end]
    i = start

    for j in range(start, end):
        if nums[j] < pivot:
            nums[i], nums[j] = nums[j], nums[i]
            i += 1

    nums[i], nums[end] = nums[end], nums[i]

    quicksort(nums, start, i - 1)
    quicksort(nums, i + 1, end)


def sort_array(nums):
    if not nums:
        return nums

    quicksort(nums, 0, len(nums) - 1)

    return nums
nums = [5, 2, 3, 1]
sorted_nums = sort_array(nums)
print(sorted_nums)

````

````python
<aside>
ðŸ’¡ 4. **Move all zeroes to end of array**

Given an array of random numbers, Push all the zeroâ€™s of a given array to the end of the array. For example, if the given arrays is {1, 9, 8, 4, 0, 0, 2, 7, 0, 6, 0}, it should be changed to {1, 9, 8, 4, 2, 7, 6, 0, 0, 0, 0}. The order of all other elements should be same. Expected time complexity is O(n) and extra space is O(1).
````

````python
def move_zeroes(nums):
    i = 0
    for j in range(len(nums)):
        if nums[j] != 0:
            nums[i], nums[j] = nums[j], nums[i]
            i += 1

    while i < len(nums):
        nums[i] = 0
        i += 1

    return nums


# Given input array
nums = [1, 9, 8, 4, 0, 0, 2, 7, 0, 6, 0]

# Move zeroes to the end of the array
modified_nums = move_zeroes(nums)

# Print the modified array
print(modified_nums)
````

````python
 5. **Rearrange array in alternating positive & negative items with O(1) extra space**

Given anÂ **array of positive**Â andÂ **negative numbers**, arrange them in anÂ **alternate**Â fashion such that every positive number is followed by a negative and vice-versa maintaining theÂ **order of appearance**.Â The number of positive and negative numbers need not be equal. If there are more positive numbers they appear at the end of the array. If there are more negative numbers, they too appear at the end of the array.

**Examples:**

> Input: Â arr[] = {1, 2, 3, -4, -1, 4}
Output: arr[] = {-4, 1, -1, 2, 3, 4}

Input: Â arr[] = {-5, -2, 5, 2, 4, 7, 1, 8, 0, -8}
Output: arr[] = {-5, 5, -2, 2, -8, 4, 7, 1, 8, 0}
> 
</aside>
````

````python
def rearrange_array(arr):
    i = 0
    j = 0
    n = len(arr)

    while i < n and j < n:
        if arr[i] < 0:
            while j < n and arr[j] < 0:
                j += 1

            if j < n:
                arr[i], arr[j] = arr[j], arr[i]

        i += 1

    return arr


# Given input array
arr = [1, 2, 3, -4, -1, 4]

# Rearrange the array
rearranged_arr = rearrange_array(arr)

# Print the rearranged array
print(rearranged_arr)
````

````python
6. Merge two sorted arrays**

Given two sorted arrays, the task is to merge them in a sorted manner.

**Examples:**

> Input: arr1[] = { 1, 3, 4, 5}, arr2[] = {2, 4, 6, 8}Â 
Output: arr3[] = {1, 2, 3, 4, 4, 5, 6, 8}

Input: arr1[] = { 5, 8, 9}, arr2[] = {4, 7, 8}
Output: arr3[] = {4, 5, 7, 8, 8, 9}
> 
````

````python
def merge_arrays(arr1, arr2):
    merged_arr = []
    i = 0
    j = 0

    while i < len(arr1) and j < len(arr2):
        if arr1[i] <= arr2[j]:
            merged_arr.append(arr1[i])
            i += 1
        else:
            merged_arr.append(arr2[j])
            j += 1

    while i < len(arr1):
        merged_arr.append(arr1[i])
        i += 1

    while j < len(arr2):
        merged_arr.append(arr2[j])
        j += 1

    return merged_arr


# Given input arrays
arr1 = [1, 3, 4, 5]
arr2 = [2, 4, 6, 8]

# Merge the arrays
merged_arr = merge_arrays(arr1, arr2)

# Print the merged array
print(merged_arr)

````

````python
 7. **Intersection of Two Arrays**

Given two integer arraysÂ `nums1`Â andÂ `nums2`, returnÂ *an array of their intersection*. Each element in the result must beÂ **unique**Â and you may return the result inÂ **any order**.

**Example 1:**

```
Input: nums1 = [1,2,2,1], nums2 = [2,2]
Output: [2]

```

**Example 2:**

```
Input: nums1 = [4,9,5], nums2 = [9,4,9,8,4]
Output: [9,4]
Explanation: [4,9] is also accepted.

```

**Constraints:**

- `1 <= nums1.length, nums2.length <= 1000`
- `0 <= nums1[i], nums2[i] <= 1000`
</aside>
````

````python
def intersection(nums1, nums2):
    nums1_set = set(nums1)
    result = []

    for num in nums2:
        if num in nums1_set:
            result.append(num)
            nums1_set.remove(num)

    return result


# Given input arrays
nums1 = [1, 2, 2, 1]
nums2 = [2, 2]

# Find the intersection
intersect = intersection(nums1, nums2)

# Print the intersection
print(intersect)

# Output: [2]
````

````python
8. **Intersection of Two Arrays II**

Given two integer arraysÂ `nums1`Â andÂ `nums2`, returnÂ *an array of their intersection*. Each element in the result must appear as many times as it shows in both arrays and you may return the result inÂ **any order**.

**Example 1:**

```
Input: nums1 = [1,2,2,1], nums2 = [2,2]
Output: [2,2]

```

**Example 2:**

```
Input: nums1 = [4,9,5], nums2 = [9,4,9,8,4]
Output: [4,9]
Explanation: [9,4] is also accepted.

```

**Constraints:**

- `1 <= nums1.length, nums2.length <= 1000`
- `0 <= nums1[i], nums2[i] <= 1000`
````

````python
def intersect(nums1, nums2):
    frequency = {}
    result = []

    # Update frequency of elements in nums1
    for num in nums1:
        frequency[num] = frequency.get(num, 0) + 1

    # Find intersection based on frequency
    for num in nums2:
        if num in frequency and frequency[num] > 0:
            result.append(num)
            frequency[num] -= 1

    return result


# Given input arrays
nums1 = [1, 2, 2, 1]
nums2 = [2, 2]

# Find the intersection
intersect = intersect(nums1, nums2)

# Print the intersection
print(intersect)


````
