```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


def find_max_subtree_sum(root):
    max_sum = float('-inf')

    def post_order_traversal(node):
        nonlocal max_sum

        if node is None:
            return 0

        left_sum = post_order_traversal(node.left)
        right_sum = post_order_traversal(node.right)

        current_sum = node.value + left_sum + right_sum
        max_sum = max(max_sum, current_sum)

        return current_sum

    post_order_traversal(root)
    return max_sum


# Construct the Binary Tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

# Find the Maximum Subtree Sum
max_subtree_sum = find_max_subtree_sum(root)

print("Maximum Subtree Sum:", max_subtree_sum)
```

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


def construct_bst(arr):
    if len(arr) == 0:
        return None

    root = TreeNode(arr[0])
    queue = []
    queue.append(root)
    i = 1

    while i < len(arr):
        parent = queue.pop(0)

        left_value = arr[i]
        if left_value is not None:
            left_child = TreeNode(left_value)
            parent.left = left_child
            queue.append(left_child)

        i += 1

        if i >= len(arr):
            break

        right_value = arr[i]
        if right_value is not None:
            right_child = TreeNode(right_value)
            parent.right = right_child
            queue.append(right_child)

        i += 1

    return root


def level_order_traversal(root):
    if root is None:
        return

    queue = []
    queue.append(root)

    while queue:
        node = queue.pop(0)
        print(node.value, end=" ")

        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)


# Given level order traversal array
level_order = [7, 4, 12, 3, 6, 8, 1, 5, 10]

# Construct the BST
root = construct_bst(level_order)

# Print the BST in level order traversal
print("BST (Level Order Traversal):")
level_order_traversal(root)
```

```python
Question-3

Given an array of size n. The problem is to check whether the given array can represent the level order traversal of a Binary Search Tree or not.

Examples:

Input1 : arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output1 : Yes

For the given arr[], the Binary Search Tree is:

            7

         /Â  Â  \

       4 Â  Â  12

     /Â  \ Â  Â  /

    3 Â  6Â  8

   / Â   / Â    \

 1Â    5  Â    10

Input2 : arr[] = {11, 6, 13, 5, 12, 10}

Output2 : No

The given arr[] does not represent the level order traversal of a BST.

</aside>
```

```python
def is_level_order_bst(arr):
    if len(arr) == 0:
        return True

    queue = []
    queue.append(arr[0])
    i = 1
    min_value = float('-inf')
    max_value = float('inf')

    while i < len(arr):
        node = queue.pop(0)
        current = arr[i]

        if current < min_value or current > max_value:
            return False

        if current < node:
            max_value = node
            queue.append(current)
        else:
            min_value = node
            queue.append(current)

        i += 1

    return True


# Given level order traversal array
level_order = [7, 4, 12, 3, 6, 8, 1, 5, 10]

# Check if the array represents a valid BST
valid_bst = is_level_order_bst(level_order)

# Print the result
if valid_bst:
    print("Yes")
else:
    print("No")
```
